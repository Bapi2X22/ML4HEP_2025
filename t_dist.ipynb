```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Sample n random numbers from a poisson distribution with mean mu
def sample_poisson(mu, n):
    return np.random.poisson(mu, n)
```

```python
# X_data = sample_poisson(1.5, 1000)
```

```python
def t_value(X_data, theta_0):
    x_bar = np.mean(X_data)
    n = len(X_data)
    term_1 = np.log(theta_0)-np.log(x_bar)
    term_2 = theta_0 - x_bar
    return -2*((n*x_bar*term_1)-n*(term_2))
```

```python
X_samples = []

for i in range(1000):
    X_samples.append(sample_poisson(1.55, 1000))
```

```python
t_values = []
for x_data in X_samples:
    t = t_value(x_data, 1.5)
    t_values.append(t)
```

```python
# plot the t-values
plt.hist(t_values, bins=50, density=True)
plt.xlabel('t-value')
plt.ylabel('Density')
plt.title('Distribution of t-values')
plt.show()
```
